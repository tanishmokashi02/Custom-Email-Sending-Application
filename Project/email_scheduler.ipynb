```python
import smtplib
import schedule
import time
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from datetime import datetime
from time import sleep

# Simulate a database of emails to send
email_list = [
    {"to": "recipient1@example.com", "subject": "Subject 1", "body": "Body of email 1"},
    {"to": "recipient2@example.com", "subject": "Subject 2", "body": "Body of email 2"},
    {"to": "recipient3@example.com", "subject": "Subject 3", "body": "Body of email 3"},
    {"to": "recipient4@example.com", "subject": "Subject 4", "body": "Body of email 4"},
    {"to": "recipient5@example.com", "subject": "Subject 5", "body": "Body of email 5"},
]

# Email sending function
def send_email(to_email, subject, body):
    from_email = "your_email@gmail.com"
    password = "your_password"  # Use an app-specific password for Gmail
    
    # Prepare email
    msg = MIMEMultipart()
    msg['From'] = from_email
    msg['To'] = to_email
    msg['Subject'] = subject
    msg.attach(MIMEText(body, 'plain'))
    
    try:
        # Set up SMTP server
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(from_email, password)
        server.sendmail(from_email, to_email, msg.as_string())
        print(f"Email sent to {to_email}")
    except Exception as e:
        print(f"Failed to send email to {to_email}: {str(e)}")
    finally:
        server.quit()

# Scheduling emails at a specific time
def schedule_emails():
    for i, email in enumerate(email_list):
        # Send emails every 10 seconds to simulate scheduling
        schedule.every(10).seconds.do(send_email, to_email=email['to'], subject=email['subject'], body=email['body'])
        print(f"Scheduled email {i+1} to {email['to']} at {datetime.now()}")

# Throttling function: Limit sending to 2 emails per minute (adjustable)
def throttle_email_sending():
    emails_sent = 0
    throttle_limit = 2  # Send 2 emails per minute

    while emails_sent < len(email_list):
        schedule.run_pending()
        emails_sent += 1
        if emails_sent % throttle_limit == 0:
            print(f"Throttling: Sleeping for 60 seconds to limit email rate.")
            sleep(60)  # Throttling: wait 60 seconds before sending more emails
        sleep(1)  # Allow the scheduled tasks to run

# Scheduling emails at 09:00 AM daily
schedule.every().day.at("09:00").do(schedule_emails)

# Start the process
throttle_email_sending()
```

```python
import schedule
import time
from sendgrid import SendGridAPIClient
from sendgrid.helpers.mail import Mail

def send_email():
    # Replace with your actual SendGrid API Key
    sendgrid_api_key = "YOUR_SENDGRID_API_KEY"
    
    # Email details
    from_email = "mokashitanish2003@gmail.com"
    to_email = "mokashiahad61@gmail.com"
    subject = "Test Email"
    body = "This is a test email sent using Python and SendGrid."

    # Create the email
    message = Mail(
        from_email=from_email,
        to_emails=to_email,
        subject=subject,
        plain_text_content=body
    )

    try:
        # Send email using SendGrid
        sg = SendGridAPIClient(sendgrid_api_key)
        response = sg.send(message)
        print("Email sent successfully!")
        print(f"Status Code: {response.status_code}")
        print(f"Response Body: {response.body}")
        print(f"Headers: {response.headers}")
    except Exception as e:
        print(f"Error: {e}")

def job():
    print("Running the email job...")
    send_email()

# Schedule the job
schedule.every().day.at("09:00").do(job)

print("Scheduler started. Waiting for the scheduled time...")

while True:
    schedule.run_pending()
    time.sleep(1)
```
