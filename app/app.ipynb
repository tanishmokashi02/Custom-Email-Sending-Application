```python
from bottle import route, run, template

# Sample data to pass into the template
dashboard_data = {
    'total_emails': 150,
    'emails_pending': 30,
    'emails_scheduled': 10,
    'emails_failed': 5,
    'response_rate': 85
}

@route('/')
def home():
    # Render the HTML template with the data
    return template('dashboard', **dashboard_data)

# Run the Bottle web server
run(host='localhost', port=8080)
```
